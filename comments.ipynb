```julia
# tuple
# very efficient
# not flexible
# error-prone
model = (0.96, 0.1, 0.3, 4)
```

```julia
model[1] = 3
```

```julia
# use a dictionary
# less error-prone
```

```julia
# instead of strings we use symbols
"a"
```

```julia
:a
```

```julia
model2 = Dict(
:α => 0.3, # pair
:β => 0.96,
:γ => 4,
:δ => 0.1
)
```

```julia
model2[:α]
```

```julia
# dict are mutable
model2[:α] = 0.4
```

```julia
# doesn't have a predetermined representation in memory
isbits(model2)
```

```julia
# create a special julia-object: a struct
```

```julia
struct AModel
    α
    β
    γ
    δ
end
```

```julia
# create a model
mm = AModel(0.3, 0.96, 4, 0.1)
```

```julia
sizeof(mm) # we have an efficient object
```

```julia
# we can access fields with .
mm.α  # human friendly
```

```julia
mm[1] # it doesn't behave like a vector of parameters
```

```julia
# functions can have optional and keyword arguments
```

```julia
# c is an optional argument
function fadd(a,b,c=1)
    a + b*c
end

# this is equivalent to 
# function fadd(a,b)
#     c=1
#     a + b*c
# end
# function fadd(a,b,c)
#     a + b*c
# end
```

```julia
fadd(1,2)
```

```julia
fadd(1,2,3)
```

```julia
#kwargs after ;
function fmul(a,b ; twice=false)
    if (twice == false)
        return a + b
    else
        return a + 2*b
    end
end
```

```julia
fmul(1,4; twice=true)
```

```julia
M = zeros(3,3)
M[1,1] = 2.0 # one based index
M[1,2] = 4.0
M[1,3] = 5.0
M
```

```julia
M[:] # first index varies first
```

```julia
# generalizes to any number of dimensions
zeros(2,3,2)
```

```julia
M
```

```julia
M[:, 2] # slice notation: any row, second column
```

```julia
# M[:, size(M,2)]
M[:, end] # take last index of second dimension
```

```julia
# slice across both dimesions

M[1:2, 1:2]
```
