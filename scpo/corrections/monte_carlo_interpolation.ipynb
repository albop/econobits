```julia
sigma = 0.05
gamma = 50
```

```julia
C(e) = exp(e)
U(x) = x^(-gamma)/(-gamma)
Xi(e) = U(C(e))
```

```julia
Xi(0)
```

```julia
randn() # normal draw
sigma*randn() # draw in a normal distribution with standard deviation sigma
```

```julia
f() = Xi(randn()*sigma)    # Xi(epsilon) for a random draw of epsilon
```

```julia
N = 100000
mc_draw() = sum([  f() for n =1:1000 ] )/N     # mean for many random draws

```

```julia
mc_draw()
```

```julia
K = 1000
many_draws = [  mc_draw() for k=1:K   ] # vector comprehension
```

```julia
pl = plot()
histogram(many_draws)
```
