```julia
function example1(pos1, pos2, pos3 )
    pos1 + pos2 + pos3*2
end
example1(1,2,5)
```

```julia
function example2(p1, p2, p3; alpha=1.0)
    p1 + p2 + p3*alpha
    # keyword arguements come after ;
end
```

```julia
example2(1,2,3) # you don't need to specify all arguements
```

```julia
example2(1,2,3; alpha=2.0)
```

# macros

```julia
dump(:(g(2,2)))
```

```julia
t1 = time()
g(2,2)
t2 = time()
print(t2-t1)
```

```julia
expr = :(g(2,2))
```

```julia
timed_expr = quote
    t1 = time()
    $expr
    t2 = time()
    print(t2-t1)
end
```

```julia
eval(expr)
```

```julia
eval(timed_expr)
```

```julia
# macros call sites with @
@time g(2,2)
```

```julia
@macroexpand @time g(2,2)
```
